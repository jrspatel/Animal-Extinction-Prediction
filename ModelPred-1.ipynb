# Applying SMOTE - oversampling tech.

```python
import numpy as np
```

```python
unique, count = np.unique(y_train,return_counts= True)
print(unique, count)
```

```python
y_train_dicts = {k:v for (k,v) in zip(unique,count)}
y_train_dicts
```

```python
from imblearn.over_sampling._smote.filter import SVMSMOTE
from imblearn.over_sampling._smote import SMOTENC
from imblearn.over_sampling import SMOTE
sm = SVMSMOTE(random_state=43)
x_trains, y_trains = sm.fit_resample(x_scaled,y_train)
```

```python
unique, count = np.unique(y_trains,return_counts= True)
print(unique, count)
```

```python
y_train_dictsmote = {k:v for (k,v) in zip(unique,count)}
y_train_dictsmote
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_trains, y_trains)
y_pred = rfc.predict(x_testscaled)
```

```python
from sklearn.metrics import f1_score
print(f1_score(y_test,y_pred,average='weighted'))
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

# Hyperparameter tunning - GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([('classifier', RandomForestClassifier())])
grid_param = [
                {'classifier':[RandomForestClassifier()],
                 'classifier__n_estimators': [10,100,1000],
                 'classifier__max_depth':[5,8,15,25,30,None],
                 'classifier__min_samples_leaf':[1,2,5,10,15,100],
                 'classifier__max_leaf_nodes':[2,5,10]} ]

gridsearch = GridSearchCV(pipe, grid_param, cv=5, verbose=0, n_jobs=-1)
bestmodel = gridsearch.fit(x_trains,y_trains)
```

```python
print(bestmodel.best_estimator_)
```

```python
print(bestmodel.score(X_test,y_test))
```
